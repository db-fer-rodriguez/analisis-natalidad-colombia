```python
# Bloque 1.1: Importar librerías básicas  
import pandas as pd  # Para manejar datos  
import matplotlib.pyplot as plt  # Para gráficos  

# Configurar matplotlib para mostrar gráficos en el notebook  
%matplotlib inline
```

```python
import os

# Bloque 1.2: Cargar archivos CSV  
file_2022 = 'datos/Nacimientos-2022.csv'
file_2023 = 'datos/Nacimientos-2023.csv'

if os.path.exists(file_2022) and os.path.exists(file_2023):
	nacimientos_2022 = pd.read_csv(file_2022, encoding='latin-1')  
	nacimientos_2023 = pd.read_csv(file_2023, encoding='latin-1')  

	# Verificar carga exitosa: Muestra las primeras 3 filas de 2022  
	print("Datos 2022:")  
	display(nacimientos_2022.head(3))  

	# Hacer lo mismo para 2023  
	print("\nDatos 2023:")  
	display(nacimientos_2023.head(3))
else:
	print("One or both files do not exist. Please check the file paths.")
```

```python
# Bloque 2.1: Unir ambos años en un único DataFrame  
df = pd.concat([nacimientos_2022, nacimientos_2023], ignore_index=True)  

# Verificar unión exitosa  
print(f"Filas totales: {len(df)}")  
print("\nEjemplo de datos combinados:")  
display(df.head(3))  
```

```python
print(f"Filas 2022: {len(nacimientos_2022)}, Filas 2023: {len(nacimientos_2023)}")
```

```python
# Bloque 3.1: ¿Cuántos valores faltantes hay por columna?  
print("Valores faltantes por columna:")  
print(df.isnull().sum())  

# Pregunta de reflexión:  
# ¿Hay columnas con más del 50% de datos faltantes? (Ej.: si una columna tiene 500k nulos en 1M filas)  
```

```python
# Columnas a eliminar  
columnas_a_eliminar = [
    'OTRO_SIT', 
    'IDCLASADMI',
    'CODPTORE',
    'CODMUNRE',
    'AREA_RES'
]

df_clean = df.drop(columns=columnas_a_eliminar, errors='ignore')  # 'errors' evita errores si alguna columna no existe
```

```python
df_clean = df_clean.drop(columns='FECHA_NACM', errors='ignore')
```

```python
df_clean = df_clean.dropna(subset=['CODPRES', 'PROFESION'])
```

```python
print("Nuevas dimensiones del DataFrame:", df_clean.shape)  
print("\nValores faltantes restantes:")  
print(df_clean.isnull().sum())
```

```python
# Bloque 4.1: Estadísticas descriptivas  
print("Estadísticas básicas:")  
display(df_clean.describe())  

# Pregunta de reflexión:  
# ¿Qué observas en las columnas numéricas (ej.: PESO_NAC, EDAD_MADRE)?
```

```python
# Bloque 4.2: Nacimientos por año  
nacimientos_por_ano = df_clean['ANO'].value_counts().sort_index()  

# Gráfico de barras  
import matplotlib.pyplot as plt  
plt.figure(figsize=(8, 4))  
nacimientos_por_ano.plot(kind='bar', color='skyblue')  
plt.title('Nacimientos por Año')  
plt.xlabel('Año')  
plt.ylabel('Número de Nacimientos')  
plt.show()  

# Pregunta de reflexión:  
# ¿Ha habido un aumento o disminución en los nacimientos entre 2022 y 2023?
```

```python
# Bloque 4.3: Nacimientos por mes  
nacimientos_por_mes = df_clean['MES'].value_counts().sort_index()  

# Gráfico de barras  
plt.figure(figsize=(10, 4))  
nacimientos_por_mes.plot(kind='bar', color='orange')  
plt.title('Nacimientos por Mes')  
plt.xlabel('Mes')  
plt.ylabel('Número de Nacimientos')  
plt.xticks(range(12), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])  
plt.show()  

# Pregunta de reflexión:  
# ¿Hay meses con más nacimientos? ¿Podría estar relacionado con festividades o clima?
```

```python
# 1. Crear una columna 'FECHA' que represente el año y mes (ej.: 2022-01)  
df_clean['FECHA'] = pd.to_datetime(  
    df_clean['ANO'].astype(str) + '-' + df_clean['MES'].astype(str),  
    format='%Y-%m'  
)  

# 2. Agrupar por fecha y contar nacimientos  
nacimientos_mensuales = df_clean.groupby('FECHA').size().reset_index(name='TOTAL')  

# Mostrar primeras filas  
print("Nacimientos mensuales:")  
display(nacimientos_mensuales.head(3))
```

```python
# 1. Configurar figura  
plt.figure(figsize=(12, 5))  

# 2. Crear gráfico de línea  
plt.plot(  
    nacimientos_mensuales['FECHA'],  
    nacimientos_mensuales['TOTAL'],  
    marker='o',  # Puntos en cada dato  
    linestyle='-',  # Línea continua  
    color='teal'  
)  

# 3. Personalizar gráfico  
plt.title('Tendencia de Nacimientos (2022-2023)')  
plt.xlabel('Fecha')  
plt.ylabel('Nacimientos Mensuales')  
plt.grid(True, linestyle='--', alpha=0.7)  # Cuadrícula transparente  

# 4. Rotar etiquetas del eje X para mejor lectura  
plt.xticks(rotation=45)  

# 5. Mostrar  
plt.tight_layout()  # Ajusta espacios para evitar cortes  
plt.show() 
```

```python
# 1. Convertir gramos a kilogramos (para mejor interpretación)  
df_clean['PESO_KG'] = df_clean['PESO_NAC'] / 1000  

# 2. Crear histograma  
plt.figure(figsize=(10, 5))  
plt.hist(  
    df_clean['PESO_KG'],  
    bins=30,  # Número de barras  
    color='purple',  
    edgecolor='black',  
    alpha=0.7  # Transparencia  
)  

# 3. Personalizar  
plt.title('Distribución de Peso al Nacer (Kg)')  
plt.xlabel('Peso (Kg)')  
plt.ylabel('Número de Bebés')  
plt.grid(True, linestyle='--', alpha=0.3)  

# 4. Líneas de referencia  
plt.axvline(df_clean['PESO_KG'].mean(), color='red', linestyle='--', label='Media')  
plt.axvline(df_clean['PESO_KG'].median(), color='green', linestyle='--', label='Mediana')  

plt.legend()  
plt.show() 
```

```python
# 1. Crear histograma de edad de la madre  
plt.figure(figsize=(10, 5))  
plt.hist(  
    df_clean['EDAD_MADRE'],  
    bins=20,  # Número de barras  
    color='salmon',  
    edgecolor='black',  
    alpha=0.7  # Transparencia  
)  

# 2. Personalizar gráfico  
plt.title('Distribución de Edad de la Madre')  
plt.xlabel('Edad de la Madre')  
plt.ylabel('Número de Madres')  
plt.grid(True, linestyle='--', alpha=0.3)  

# 3. Líneas de referencia  
plt.axvline(df_clean['EDAD_MADRE'].mean(), color='blue', linestyle='--', label='Media')  
plt.axvline(df_clean['EDAD_MADRE'].median(), color='green', linestyle='--', label='Mediana')  

plt.legend()  
plt.show()
```

```python
# 1. Crear gráfico de dispersión  
plt.figure(figsize=(10, 5))  
plt.scatter(  
    df_clean['EDAD_MADRE'],  
    df_clean['PESO_KG'],  
    alpha=0.5,  # Transparencia para ver densidad  
    color='darkblue'  
)  

# 2. Personalizar gráfico  
plt.title('Relación entre Edad de la Madre y Peso al Nacer')  
plt.xlabel('Edad de la Madre')  
plt.ylabel('Peso al Nacer (Kg)')  
plt.grid(True, linestyle='--', alpha=0.3)  

# 3. Líneas de referencia  
plt.axhline(df_clean['PESO_KG'].mean(), color='red', linestyle='--', label='Media de Peso')  
plt.axvline(df_clean['EDAD_MADRE'].mean(), color='blue', linestyle='--', label='Media de Edad')  

plt.legend()  
plt.show() 
```
